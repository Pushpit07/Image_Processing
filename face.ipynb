```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

import cv2 # computer vision
```

```python
frame = cv2.imread("thor.png")
```

```python
plt.imshow(frame) # cv2 uses BGR instead of RGB
```

```python
rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
!wget https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml -P drive/gaze
```

```python
cascades_path = 'drive/gaze/'
```

```python
classifier = cv2.CascadeClassifier(cascades_path + "haarcascade_frontalface_default.xml")
```

```python
classifier.detectMultiScale(frame)
```

```python
faces = classifier.detectMultiScale(frame)
```

```python
face = faces[0]
```

```python
x, y, w, h = face
```

```python
face = rgb[y:y+h, x:x+w]
```

```python
plt.imshow(face)
```

```python
out = cv2.rectangle(rgb, (x, y), (x+w, y+h), (0,0,255), 3)
```

```python
plt.imshow(out)
```
